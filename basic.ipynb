```python
a = int(input('enter a value:'))
b = int(input('enter a value:'))
op = input('enter operator:')
if op == '+':
    print('addition:',a+b)
elif op == '-':
    print('subtraction:',a-b)
elif op == '*':
    print('multiplication:',a*b)
elif op == '/':
    print('division:',a/b)
elif op == '//':
    print('exponent:',a//b)
elif op == '%':
    print('modulus:',a%b)
elif op == '**':
    print('power:',a**b)
else:
    print('invalid')
```

```python
#to print factorial of a number using recursion
def fact(num):
    if num<=0:
        return 1
    else:
        return num*fact(num-1)
```
